```python
import numpy as np

%matplotlib  inline

import matplotlib.pyplot as plt
import matplotlib
```

```python
data = np.load('train_loss.npz')

x=range(len(data['L']))
y=data['L'].reshape(-1)

plt.title('train last loss cost : %f' %y[-1])
plt.plot(x,y)
plt.xlabel('epoch*batch')
plt.ylabel('loss')
plt.show()

print ('average : '+str(np.average(data['L'])))
print ('min : '+str(np.min(data['L'])))
```
